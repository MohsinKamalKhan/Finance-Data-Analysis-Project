```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
```

```python
start = datetime.datetime(2006, 1, 1)  # year, month, date
end  = datetime.datetime(2016, 1, 1)

# Bank of America
BAC = pd.read_pickle('all_banks.pkl')['BAC']
# CitiGroup
C = pd.read_pickle('all_banks.pkl')['C']
# Goldman Sachs
GS = pd.read_pickle('all_banks.pkl')['GS']
# JPMorgan Chase
JPM = pd.read_pickle('all_banks.pkl')['JPM']
# Morgan Stanley
MS = pd.read_pickle('all_banks.pkl')['MS']
# Wells Forgo
WFC = pd.read_pickle('all_banks.pkl')['WFC']

WFC
```

```python
# Create a list of the ticker symbols (as strings) in alphabetical order.
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
# use pd.concat to concatenate the bank dataframes together in single dataframe
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], axis=1, keys=tickers)
bank_stocks
```

```python
# Set the column name for Banks
bank_stocks.columns.names=['Bank Ticker', 'Stock Info']
bank_stocks.head()
```

```python
# What is the max Close price for each bank's stock throughout the time period
for tick in tickers:
    print('Max Closing Pricing for ' + tick + ' bank is ' + str(bank_stocks[tick]['Close'].max()))

# or

bank_stocks.xs(key='Close', axis=1, level='Stock Info').max()
```

```python
# Create a new empty DataFrames called returns

returns = pd.DataFrame()
```

```python
'''
    use pandas pct_change() method on Close column of each ticker to create column of return values.
'''

for tick in tickers:
    returns[tick + ' Return'] = bank_stocks[tick]['Close'].pct_change()

returns.head()
```

```python
# Create a pairplot using seaborn of the returns dataframe. What stock stands out to you?
import seaborn as sns
sns.pairplot(returns[1:])
```

```python
# using returns Dataframe figure out what dates each bank stock had best and worst single day returns
print(returns.idxmin()) # return index of minimum value for each column
print(returns.idxmax()) # return index of maximum value for each column
```

```python
# Take a look at the standard deviation of the returns, which stock would you classify as the riskiest over the entire time period?

# High standard deviation => a lot of changes (risky)
# Low standard deviation => stability (not risky)

returns.std()
```

```python
# Which would you classify as the riskiest for the year 2015?
returns.loc['2015-01-01':'2015-12-31'].std()
```

```python
# Create a distplot using seaborn of 2015 returns of Morgan Stanley

sns.distplot(returns.loc['2015-01-01':'2015-12-31']['MS Return'], color='green', bins=50)
```

```python
# Create a distplot using seaborn of 2008 returns for CitiGroup
sns.distplot(returns.loc['2008-01-01':'2008-12-31']['C Return'], color='red', bins=50)
```

```python
# Create a line plot showing Close price for each bank for the entire index of time.
import matplotlib.pyplot as plt
for tick in tickers:
    bank_stocks[tick]['Close'].plot(label=tick, figsize=(12, 4))
plt.legend()
```
